## 1) Pêndulo simples, mas além das oscilações de pequenas amplitudes.

Resolva a equação de movimento do pêndulo com as condições de contorno abaixo:

$$ \dfrac{\partial ^2 x}{\partial t^2} = -\omega_0 ^2 \sin(x) - b.v + F.cos(\omega_e t) $$
sendo $x(0)=x_0$ e $v(0)=0$.

Use sempre $\omega_0 = 2\pi$. Os outros parâmetros e a posição inicial $0<x_0<\pi$, nós iremos variar. Em todos os casos abaixo, faça gráficos da trajetória (x vs t) e do espaço de fase (x vs v).

##### a) Use a técnica de redução de ordem para reescrever (no papel mesmo) esta EDO na forma de duas EDOs de primeira ordem. Para isso introduza a variável $v(t)=\partial x/\partial t$. Rode para tempos no intervalo $0<t<t_{max}$, e escolha um valor razoável para $t_{max}$ tal que veja várias oscilações claramente.

Para reduzir a ordem da EDO, iremos usar $v(t)=\partial x/\partial t$, ficando assim com $ \dfrac{\partial v}{\partial t} = -\omega_0 ^2 \sin(x) - b.v + F.cos(\omega_e t) $

Com essas duas EDOs, podemos implementar o código:

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Condições iniciais
x0 = np.pi
v0 = 0
ome = 2*np.pi
b=1 #Valores aleátórios de b, F e omega e
F=1
ome_e=1

#Definindo a função
def f(t, y):
    v = y[0]
    x = y[1]
    dvdt = -(ome **2)*np.sin(x) - b*v + F*np.cos(ome_e *t)
    dxdt = v
    return np.array([dvdt, dxdt])

#Solução da EDO
tlist=np.linspace(0, 10, 1000)
sol = solve_ivp(f, [0, 10], (v0, x0), method='RK45', t_eval=tlist)
t=sol.t
v=sol.y[0]
x=sol.y[1]

#Plotando o gráfico
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1) 
plt.plot(t, x)
plt.title('x vs t')
plt.xlabel('t')
plt.ylabel('x')

plt.subplot(1, 2, 2) 
plt.plot(v, x, color='orange')
plt.title('v vs x')
plt.xlabel('x')
plt.ylabel('v')

plt.show()
```

## 2) Trajetória de uma bola girando

Durante a trajetória a bola com velocidade está sob ação das forças
de gravidade $F_g$, de arrasto do ar $F_d$ e força de Magnus $F_m$ devido à
rotação. A figura ao lado ilustra a rotação com frequência $\omega$ e as forças de
arrasto e de Magnus. Estas podem ser escritas como:

$F_g = -mg \hat z$

$F_d=-mf(v)v^2 \hat v = -mf(v)v\hat v$, sendo $f(v)=0.0039+\frac{0.0058}{1+exp[(v-v_d)/\Delta]}$

$F_m=S(v)\vec \omega \times \hat v$, sendo $S(v)=(4.1\times10^{-4})m$

Acima, note que $v=|\vec v|$e a massa m da bola sempre irá se cancelar quando igualar a resultante de
forças à F = ma. Na força de arrasto a função f(v) é empírica e descreve como a força de resistência do ar varia com a velocidade da bola. Os parâmetros experimentais para bolas de baseball são $v_d$= 35 m/s e $\Delta$= 5 m/s. Na força de Magnus, a função S(v) também é empírica. O valor constante é uma aproximação válida para bolas de baseball.

Considerando:

-O lançamento é na direção x e parte da altura do ombro do jogador (eixo z aponta para cima):
$\vec v(0)=v_0 \hat x$, $v_0$ de 20 a 50 m/s.
$\vec r(0)=z_0 \hat z$, podemos aproximar $z_0 =2$m. 

-A rotação da bola será constante e na direção y (fastball), $\vec \omega = \omega _0 \hat y$, com $\omega_0$ até 2000 rpm.

Calcule a trajetória da bola para:

##### a) $\omega _0 = 0$

Primeiramente, decompondo para x e z:

Em z: $\vec F_{res_z}=\vec F_g +\vec F_{dz} + \vec F_{mz} = - mg \hat z + m f(v_z) v_z^2 \hat z + S(v_z)\omega _0 v_x \hat z$

Em x: $\vec F_{res_x}=\vec F_{dx} + \vec F_{mx} = - m f(v_x) v_x^2 \hat x + S(v_x)\omega _0 v_x \hat x$

Logo sendo $\vec F=m \vec a$, temos as duas EDOs:

$$\frac{\partial ^2 \vec z}{\partial t^2}=[-g+f(v_z)v_z^2+S(v_z)\omega _0 v_x]\hat x$$
$$\frac{\partial ^2 \vec x}{\partial t^2}=[-f(v_x)v_x^2+S(v_x)\omega _0 v_z]\hat x$$

Tendo as duas EDOs, podemos montar o código:

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Condições iniciais
x0 = 0
v0 = 50 #Considerando a velocidade inicial em x de 50 m/s
z0 = 2 #Altura do jogador em metros
u0 = 0 #Velocidade inicial em z
S = 0.00041
g = 9.8 #Acelaração gravitacional em m/s^2
omega = 0

#Definindo a função separando em nas direções x e z
def f(t, y):
    v = y[0]
    x = y[1]
    u = y[2]
    z = y[3]
    dvdt = -(v**2)*(0.0039+0.0058/(1+np.exp((v-35)/5))) + S*omega*u
    dxdt = v
    dudt = -g +(u**2)*(0.0039+0.0058/(1+np.exp((u-35)/5))) + S*omega*v
    dzdt = u
    return np.array([dvdt, dxdt, dudt, dzdt])

#Solução da EDO
tlist=np.linspace(0, 10, 1000)
sol = solve_ivp(f, [0, 10], (v0, x0, u0, z0), method='RK45', t_eval=tlist)
z=sol.y[3]
x=sol.y[1]

#Plotando o gráfico
plt.plot(x, z)
plt.title('z vs x')
plt.ylim(0, 2.5)
plt.xlim(0, 40)
plt.xlabel('x')
plt.ylabel('z')
plt.show()
```

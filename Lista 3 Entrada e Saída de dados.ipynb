## 1) Escrever um código que gere os vetores (arrays) abaixo e os salve em arquivos.

```python
import numpy as np

# Criando os arrays
x = np.linspace(0, 2*np.pi, 15)
print('O vetor x é:', x)
y= np.sin(x)
print('O seno do vetor x é:', y)
z= np.cos(x)
print('O cosseno do vetor x é:', z)

# Criando os arquivos e salvando os arrays
arquivo1 = open('arquivo1.txt', "wb")
np.save(arquivo1, x)
arquivo1.close()

arquivo2 = open('arquivo2.txt', "wb")
np.save(arquivo2, y)
arquivo1.close()

arquivo3 = open('arquivo3.txt', "wb")
np.save(arquivo3, z)
arquivo1.close()
```

#### a) Escreva estes vetores em um arquivo “dados.txt” usando o comando np.savetxt(...). Como controlar se os dados serão salvos em linhas (x,y,z) ou colunas (x,y,z)?


```python
import numpy as np

# Criando os arrays
x = np.linspace(0, 2*np.pi, 15)
y= np.sin(x)
z= np.cos(x)

# Criando o arquivo dados.txt e salvando em colunas
np.savetxt('dados.txt', np.c_[x, y, z])
vetores = np.loadtxt('dados.txt')
vetores
```

#### b) Escreva estes vetores em um arquivo binário “dados.npy” usando o np.save(...).

```python
import numpy as np

# Criando os arrays
x = np.linspace(0, 2*np.pi, 15)
y= np.sin(x)
z= np.cos(x)

# Criando o arquivo dados.npz
dados = open('dados.npz', "wb")
np.savez('dados.npz', vetorx=x, vetory=y, vetorz=z)
vetores = np.load('dados.npz')

xvet = vetores['vetorx']
print('O vetor x é:', xvet)
yvet = vetores['vetory']
print('O vetor y é:', yvet)
zvet = vetores['vetorz']
print('O vetor z é:', zvet)
```

## 2) Escreva outros códigos que leia estes arquivos e faça os gráficos de y vs x e z vs x. Sendo um código para o caso (a) dados.txt e outro para (b) dados.npy.

##### a) dados.txt

```python
import numpy as np
import matplotlib.pyplot as plt

# Puxando o arquivo dados.txt
vetores = np.loadtxt('dados.txt')

xlado = vetores[:,[0]]
ylado = vetores[:,[1]]
zlado = vetores[:,[2]]

#Plotagem do gráfico
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1) 
plt.plot(xlado, ylado, label='y = seno(x)')
plt.title('Gráfico do seno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xlado, zlado, label='y = cos(x)', color='orange')
plt.title('Gráfico do cosseno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

##### b) dados.npz

```python
import numpy as np
import matplotlib.pyplot as plt

#Puxando o arquivo dados.npz
vetores = np.load('dados.npz')

xlado = vetores['vetorx']
ylado = vetores['vetory']
zlado = vetores['vetorz']

#Plotagem do gráfico
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1) 
plt.plot(xlado, ylado, label='y = seno(x)')
plt.title('Gráfico do seno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xlado, zlado, label='y = cos(x)', color='orange')
plt.title('Gráfico do cosseno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

#### a) Qual é a diferença entre os dois tipos de arquivo?

Arquivos .txt possuem uma velocidade de leitura muito infeiror aos arquivos .npy, isso ocorre devido à estrutura do arquivo .npy binário padrão que apresenta uma única matriz NumPy arbitrária no disco, sendo projetado para ser o mais limpo e puro possível, por isso ele ocupa menos espaço e é lido mais rapidamente.


#### b) Quando devemos usar o np.savetxt ou o np.save?

O np.savetxt deve ser utilizado apenas para salvar arrays em formato txt, logo ele é interessante para salvar textos para serem lidos manualmente e utilizar em outros programas. Já o np.save deve ser utilizado para salvar um array em um arquivo binário no formato numpy (.npy).

### 1) Implemente uma função semelhante ao exemplo acima, mas utilizando o método de Runge-Kutta de 4a ordem.

Método de Runge-Kutta para exponencial de t

```python
import numpy as np

# Função EDO
def f(y, t):
    return np.exp(t)

# Método de Runge-kutta
def runge(f, y0, dt):
    N = len(t)
    dt = t[1]-t[0]
    y = np.zeros(N)
    y[0] = y0
    for j in range(N-1):
        k1 = f(y[j],t[j])
        k2 = f(y[j]+0.5*k1*dt, t[j]+0.5*dt)
        k3 = f(y[j]+0.5*k2*dt, t[j]+0.5*dt)
        k4 = f(y[j]+k3*dt, t[j]+dt)
        y[j+1] = y[j] + dt*(k1+2*k2+2*k3+k4)/6
    return y

# Condições
t = np.arange(0, 10, 0.1)

# Plotagem do gráfico
y = runge(f, 1, t)
plt.plot(t, y, label='Runge-Kutta')
plt.plot(t, np.exp(t), '--', label='Exato')
plt.ylim(0,3000)
plt.legend()
plt.xlabel('t')
plt.ylabel('y(y)')
plt.show()
```

### 2) Use sua função para resolver a mesma EDO vista em aula e compare o resultado com o método de Euler para um mesmo . Varie para ver como Euler e Runge-Kutta se comportam.

Comparando o método de Euler com o de Runge-kutta para a função exponencial:

```python
import numpy as np

# Função EDO
def f(y, t):
    return np.exp(t)

# Método de Euler
def euler(f, y0, t):
    N = len(t)
    dt = t[1]-t[0]
    y = np.zeros(N)
    y[0] = y0
    for j in range(N-1):
        y[j+1] = y[j]+ f(y[j], t[j])*dt
    return y

# Método de Runge-kutta
def runge(f, y0, dt):
    N = len(t)
    dt = t[1]-t[0]
    y = np.zeros(N)
    y[0] = y0
    for j in range(N-1):
        k1 = f(y[j],t[j])
        k2 = f(y[j]+0.5*k1*dt, t[j]+0.5*dt)
        k3 = f(y[j]+0.5*k2*dt, t[j]+0.5*dt)
        k4 = f(y[j]+k3*dt, t[j]+dt)
        y[j+1] = y[j] + dt*(k1+2*k2+2*k3+k4)/6
    return y

# Condições
t = np.arange(0, 10, 0.1)

# Plotagem do gráfico
y1 = runge(f, 1, t)
y2 = euler(f, 1, t)
plt.plot(t, y1, label='Runge-Kutta')
plt.plot(t, y2, label='Euler')
#plt.plot(t, np.exp(t), '--', label='Exato')
plt.legend()
plt.xlabel('t')
plt.ylabel('y(y)')
plt.show()
```
